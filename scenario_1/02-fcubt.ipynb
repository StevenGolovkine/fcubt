# Perform fCUBT on the data

```python
# Load packages
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pickle

from FDApy.clustering.fcubt import Node, FCUBT
from FDApy.representation.functional_data import DenseFunctionalData
from FDApy.visualization.plot import plot

from matplotlib import colors as mcolors
COLORS = ['#377eb8', '#ff7f00', '#4daf4a',
          '#f781bf', '#a65628', '#984ea3',
          '#999999', '#e41a1c', '#dede00']

#matplotlib.use("pgf")
#matplotlib.rcParams.update({
#    "pgf.texsystem": "pdflatex",
#    'font.family': 'serif',
#    'text.usetex': True,
#    'pgf.rcfonts': False,
#})
```

```python
# Load data
with open('./data/scenario_1.pkl', 'rb') as f:
    data_fd = pickle.load(f)
```

```python
# Build the tree
root_node = Node(data_fd, is_root=True)
fcubt = FCUBT(root_node=root_node)
```

```python
# Growing
fcubt.grow(n_components=0.95, min_size=10)
```

```python
# Joining
fcubt.join(n_components=0.95)
```

```python
# Plot the tree
fig = plt.figure(figsize=(10, 10), constrained_layout=True)
fcubt.plot(fig=fig)
plt.savefig('./figures/maximal_tree.pdf')
```
